```python
import pandas as pd
from pathlib import Path

task_str = "Pick up the box on the conveyor belt and place it in the container"

# 把 task 放到 index 里，列里只保留 task_index
df = pd.DataFrame(
    {"task_index": [0]},
    index=[task_str],
)

output_path = Path("./test.parquet")
df.to_parquet(output_path)  # 注意这里不要加 index=False，让 index 一起存进去

print(f"Written to {output_path}")
```
